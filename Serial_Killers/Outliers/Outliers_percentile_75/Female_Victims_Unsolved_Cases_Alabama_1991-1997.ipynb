# Female Victims. Unsolved Cases. Alabama 1991-1997

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from mlxtend.preprocessing import minmax_scaling

from jupyterthemes import jtplot
jtplot.style(figsize=(15, 9))
```

## Data

```python
map_data = pd.read_csv('../data/SHR65_22.csv', index_col=0)
map_data.head()
```

```python
df = map_data[(map_data['VicSex'] == 'Female')
             & (map_data['Solved'] == 'No')
             & (map_data['State'] == 'Alabama')
             & (map_data['Year'].between(1991, 1997))]
df.head()
```

```python
df.groupby('CNTYFIPS').count()['Ori']
```

```python
fig, ax = plt.subplots(figsize=(21, 17))

df.groupby('CNTYFIPS').count().sort_values(by='Ori', ascending=True)['Ori'].plot(kind='barh', ax=ax)
ax.set_title('Cases by City')
```

```python
fig, ax = plt.subplots(3, 1, figsize=(21, 15))

df.groupby('VicRace').count().sort_values(by='Ori', ascending=True)['Ori'].plot(kind='barh', ax=ax[0])
ax[0].set_title('Cases by Victims Race')

df.groupby('Weapon').count().sort_values(by='Ori', ascending=True)['Ori'].plot(kind='barh', ax=ax[1])
ax[1].set_title('Cases by Weapon')

df.groupby('VicAge').count()['Ori'].plot(kind='bar', ax=ax[2])
ax[2].set_title('Cases by Victims Age')
```

```python
jefferson = df[df['CNTYFIPS'] == 'Jefferson, AL']   # population: 665_409 !Jefferson County!
mobile = df[df['CNTYFIPS'] == 'Mobile, AL']         # population: 183_289
montgomery = df[df['CNTYFIPS'] == 'Montgomery, AL'] # population: 196_986
```

```python
jefferson.shape[0], mobile.shape[0], montgomery.shape[0]
```

```python
fig, ax = plt.subplots(1, 3, figsize=(21, 9))

jefferson.groupby('Year').count()['Ori'].plot(kind='barh', ax=ax[0]).invert_yaxis()
ax[0].set_title('Jefferson, AL')

mobile.groupby('Year').count()['Ori'].plot(kind='barh', ax=ax[1]).invert_yaxis()
ax[1].set_title('Mobile, AL')

montgomery.groupby('Year').count()['Ori'].plot(kind='barh', ax=ax[2]).invert_yaxis()
ax[2].set_title('Montgomery, AL')
```

```python
fig, ax = plt.subplots(3, 1, figsize=(11, 15))

jefferson.groupby('Weapon').count().sort_values(by='Ori', ascending=True)['Ori'].plot(kind='barh', ax=ax[0])
ax[0].set_title('Jefferson, AL')

mobile.groupby('Weapon').count().sort_values(by='Ori', ascending=True)['Ori'].plot(kind='barh', ax=ax[1])
ax[1].set_title('Mobile, AL')

montgomery.groupby('Weapon').count().sort_values(by='Ori', ascending=True)['Ori'].plot(kind='barh', ax=ax[2])
ax[2].set_title('Montgomery, AL')
```

```python
j = jefferson.groupby(['Weapon', 'VicRace']).count()['Ori'].to_frame().reset_index()
j = pd.pivot(j, index='Weapon', columns='VicRace', values='Ori')

mb = mobile.groupby(['Weapon', 'VicRace']).count()['Ori'].to_frame().reset_index()
mb = pd.pivot(mb, index='Weapon', columns='VicRace', values='Ori')

mg = montgomery.groupby(['Weapon', 'VicRace']).count()['Ori'].to_frame().reset_index()
mg = pd.pivot(mg, index='Weapon', columns='VicRace', values='Ori')
```

```python
fig, ax = plt.subplots(3, 1, figsize=(11, 15))

j.sort_values(by='Black').plot(kind='barh', ax=ax[0])
ax[0].set_title('Jefferson, AL')

mb.sort_values(by='Black').plot(kind='barh', ax=ax[1])
ax[1].set_title('Mobile, AL')

mg.sort_values(by='Black').plot(kind='barh', ax=ax[2])
ax[2].set_title('Montgomery, AL')
```
