```python
df = map_data[(map_data['Weapon']=='Strangulation - hanging')
              & (map_data['State']=='Texas')
              & (map_data['VicSex']=='Female')
              & (map_data['Year'].between(1976, 1981))]
df.head()
```

```python
norm = map_data[(map_data['Weapon']=='Strangulation - hanging')
                & (map_data['State']=='Texas')
                & (map_data['VicSex']=='Female')
                & (~map_data['Year'].between(1976, 1981))]

norm.groupby('Solved').mean()
```
